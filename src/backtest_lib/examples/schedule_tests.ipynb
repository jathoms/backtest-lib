```python
import datetime as dt

import backtest_lib.backtest.schedule as sch

sched = sch.make_decision_schedule([dt.datetime(2025, 1, 1), dt.datetime(2025, 1, 2)])
for period in sched:
    print(period)
```

```python
# ensure re-iterability (sched is not consumed by initial iteration)
for period in sched:
    print(period)
```

```python
cron_sched = sch.make_decision_schedule("0 * * * *", start=dt.datetime(2025, 2, 1))
cron_sched_iter = iter(cron_sched)

for _ in range(10):
    next_period = next(cron_sched_iter)
    print(next_period)
```

```python
interval_sched = sch.make_decision_schedule("2h", start=dt.datetime(2021, 11, 13))
interval_sched_iter = iter(interval_sched)

for _ in range(15):
    next_period = next(interval_sched_iter)
    print(next_period)
```

```python
bounded_schedule = sch.make_decision_schedule(
    "1w",
    start=dt.datetime(2024, 2, 1),
    end=dt.datetime(2024, 2, 3),
)

for period in bounded_schedule:
    print(period)
```
